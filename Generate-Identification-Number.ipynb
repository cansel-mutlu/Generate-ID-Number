```python
print("""##################################################
This program generates id number after user input.
##################################################""")

import random
def generate_id_number():
    
    # Create random number with 10 digits
    
    rand_num_with_10_digits = random.randint((10**9),((10**10)-1))

    # Get even indexes from 0 to 9 

    digits_from_index_0_to_index_9 = str(rand_num_with_10_digits)[0:9:2]
    
    # Get odd indexes from 1 to 8
    
    digits_from_index_1_to_index_8 = str(rand_num_with_10_digits)[1:8:2]
    
    # Get sum of the first 10 digits
    
    sum_of_the_first_10_digits = 0
    for digit in str(rand_num_with_10_digits):
        sum_of_the_first_10_digits += int(digit)
    
    # Create a variable for last digit of id number and assign this variable to the last digit of sum of the first 10 digits
        
    last_digit_of_id = str(sum_of_the_first_10_digits)[-1]
    
    # Get sum of the digits from index 0 to index 9

    sum_of_the_digits_from_index_0_to_index_9 = 0
    for digit in digits_from_index_0_to_index_9:
        sum_of_the_digits_from_index_0_to_index_9 += int(digit)
        
    multiplied_by_7 = (sum_of_the_digits_from_index_0_to_index_9) * 7
    multiplied_by_8 = (sum_of_the_digits_from_index_0_to_index_9) * 8
    
    # Create a second variable for last digit of id number and assign this variable to the last digit of number multiplied by 8
    
    last_digit_of_id_2 = str(multiplied_by_8)[-1]

    # Get sum of the digits from index 1 to index 8
    
    sum_of_the_digits_from_index_1_to_index_8 = 0
    for digit in digits_from_index_1_to_index_8:
        sum_of_the_digits_from_index_1_to_index_8 += int(digit)
        
    multiplied_by_9 = (sum_of_the_digits_from_index_1_to_index_8) * 9

    sum_of_the_multiplied_sum = multiplied_by_7 + multiplied_by_9
    
    # Create a variable for tenth digit of id number and assign this variable to the last digit of sum of the multiplied sum

    tenth_digit_of_id = str(sum_of_the_multiplied_sum)[-1]


    if last_digit_of_id == last_digit_of_id_2:
        
        # Create variables for first 9 digits, tenth digit and last digit of id
    
        number_1 = str(rand_num_with_10_digits)[0:9]  
        number_2 = tenth_digit_of_id
        number_3 = last_digit_of_id
        
        # Combine these variables for creating an id number

        id_number = number_1 + number_2 + number_3
        return id_number
    else:
        return None
    
while True:

    try:
        input_number = int(input("Enter a number:"))       
        break
    except ValueError:
        print("Please enter a positive integer")
        
# Create a list to store unique id numbers        

unique_list = []
list_length = 0

while list_length < input_number: # Use while loop for generating id number equal to user input
    r = generate_id_number()       
    
    # Add if id number not exist in the list
    
    if r not in unique_list and r is not None:  
        list_length += 1
        unique_list.append(r)
        print(unique_list) # Print id numbers list
```
